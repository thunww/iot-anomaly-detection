```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("data/raw/Train_Test_IoT_Modbus.csv")
df.head()
```

```python
df.info()
```

```python
df['type'].value_counts()
```

```python
df_numeric = df.drop(columns=["date", "time", "type"])
df_numeric.head()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df_numeric.corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap (Modbus Features)")
plt.show()
```

```python
df_numeric.drop(columns=["label"]).hist(bins=40, figsize=(12,8))
plt.suptitle("Feature Distributions (Histogram)", fontsize=16)
plt.show()
```

```python
features = ['FC1_Read_Input_Register','FC2_Read_Discrete_Value',
            'FC3_Read_Holding_Register','FC4_Read_Coil']

for f in features:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df['label'], y=df[f])
    plt.title(f"{f}: Normal vs Attack")
    plt.show()
```

```python
df['timestamp'] = pd.to_datetime(df['date'] + ' ' + df['time'])
df_sorted = df.sort_values('timestamp')

plt.figure(figsize=(12,4))
plt.plot(df_sorted['timestamp'], df_sorted['FC3_Read_Holding_Register'])
plt.title("FC3 Register Trend Over Time")
plt.xlabel("Timestamp")
plt.ylabel("Value")
plt.show()
```
